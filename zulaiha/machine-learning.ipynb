# Trend over time.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visual style using the updated method
sns.set_theme(style="whitegrid")

# Total CO₂ emissions per year
yearly_emissions = df.groupby('Year')['Co2_Emissions_MetricTons'].sum()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=yearly_emissions.index, y=yearly_emissions.values)
plt.title("Total Global CO₂ Emissions (2000–2022)")
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (Metric Tons)")
plt.tight_layout()
plt.show()
```

# Group by year and aggregate emissions for each sector

```python

sector_trends = df.groupby('Year')[
    ['Industrial_Co2_Emissions_MetricTons',
     'Automobile_Co2_Emissions_MetricTons',
     'Agriculture_Co2_Emissions_MetricTons',
     'Domestic_Co2_Emissions_MetricTons']
].sum()

# Plot
plt.figure(figsize=(12, 6))
sector_trends.plot()
plt.title("CO₂ Emissions by Sector (2000–2022)")
plt.xlabel("Year")
plt.ylabel("CO₂ Emissions (Metric Tons)")
plt.legend(title="Sector")
plt.grid(True)
plt.tight_layout()
plt.show()
```
